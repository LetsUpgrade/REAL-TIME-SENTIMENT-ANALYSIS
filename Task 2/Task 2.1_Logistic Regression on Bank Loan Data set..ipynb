```python
import pandas as pd
import statsmodels.api as sm
dataset=pd.read_excel("D:\Study\AI_ML\Assignments\Day 21\Bank_Personal_Loan_Modelling.xlsx",sheet_name=1)
dataset.columns
```

```python
dataset1=dataset.dropna()
dataset1
```

```python
dataset2=dataset1.drop_duplicates()
dataset2
```

```python
dataset3=dataset2.drop(["ID","ZIP Code"],axis=1)
dataset3
```

```python
dataset3.columns
```

```python
y=dataset3[["Personal Loan"]]
X=dataset3[['Age', 'Experience', 'Income', 'Family', 'CCAvg',
       'Education', 'Mortgage', 'Securities Account',
       'CD Account', 'Online', 'CreditCard']]
```

```python
X1=sm.add_constant(X)
```

```python
Bankloan=sm.Logit(y,X1)
```

```python
result=Bankloan.fit()
```

```python
result.summary()
```

# The Variables Income, Family , CCAvg ,Education , Securities,Account ,CD Account , Online , CreditCard are significantly important for getting the Personal loan.
